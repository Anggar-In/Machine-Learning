```
import keras_ocr
from matplotlib import pyplot as plt
```

```
# keras-ocr will automatically download pretrained
# weights for the detector and recognizer.
pipeline = keras_ocr.pipeline.Pipeline()
```

```
from google.colab import files
uploaded = files.upload()

images = [keras_ocr.tools.read(img_path) for img_path in uploaded.keys()]

# Get the predictions
prediction_groups = pipeline.recognize(images)

# Plot the predictions
fig, ax = plt.subplots(figsize=(10, 10))  # Ukuran gambar dapat disesuaikan
keras_ocr.tools.drawAnnotations(image=images[0], predictions=prediction_groups[0], ax=ax)

plt.show()
```
